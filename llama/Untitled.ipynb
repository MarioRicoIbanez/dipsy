```python
from datasets import load_dataset
import matplotlib.pyplot as plt
from collections import Counter

# Carga el dataset desde Hugging Face
dataset = load_dataset("RikoteMaster/isear_augmented")

# Obtiene las emociones y cuenta su frecuencia
emotions = dataset["train"]["Emotion"]
emotion_counts = Counter(emotions)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
emotions, counts = zip(*emotion_counts.items())
plt.bar(emotions, counts, color='skyblue')
plt.title('Cantidad de Sentencias por Emoción')
plt.xlabel('Emoción')
plt.ylabel('Cantidad de Sentencias')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
